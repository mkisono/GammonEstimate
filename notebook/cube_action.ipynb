```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
start = 19
end = 122
x = np.linspace(start, end, end-start+1)
```

```python
# 136 ページの表
def trailer_can_take(x):
    range_mapping = {
        range(19, 26): 2,
        range(26, 33): 3,
        range(33, 40): 4,
        range(40, 47): 5,
        range(47, 54): 6,
        range(54, 62): 7,
        range(62, 70): 8,
        range(70, 79): 9,
        range(79, 89): 10,
        range(89, 100): 11,
        range(100, 111): 12,
        range(111, 123): 13
    }
    
    for range_, value in range_mapping.items():
        if x in range_:
            return x + value
```

```python
# plot y = x line and y = trailer_can_take(x) line
plt.plot(x, x, label='leader pip count')
plt.plot(x, [trailer_can_take(i) for i in x], label='trailer can take pip count')
plt.legend()
plt.show()
```

```python
# 137ページの公式
def trailer_can_take_formula(x):
    if x <= 62:
        # subtract 5 from x, divede by 7, and round down
        return (x - 5) // 7 + x
    # 10% round up to the nearest integer and add 1
    return int(np.ceil(x * 0.1)) + 1 + x
```

```python
# compare trailer_can_take and trailer_can_take_formula
y = [trailer_can_take(x) for x in x]
y_formula = [trailer_can_take_formula(x) for x in x]
plt.plot(x, y, label="trailer_can_take")
plt.plot(x, y_formula, label="trailer_can_take_formula")
plt.legend()
plt.show()
```
