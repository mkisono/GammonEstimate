```python
import pandas as pd
```

```python
df = pd.read_pickle('../data/xg_data_all.pkl')
```

```python
df.info()
```

```python
# remove rows with ActiveP == 0
df = df[df['ActiveP'] != 0]
# remove rows with invalid cube
df = df[df['Doubled_equB'] != 0.0]
```

```python
df.loc[df['ActiveP'] == -1, 'Position'] = df.loc[df['ActiveP'] == -1, 'Position'].apply(lambda x: [-i for i in x[::-1]])
```

```python
df['player_pip_count'] = df['Position'].apply(lambda x: sum(p * i for i, p in enumerate(x) if p > 0))
df['opponent_pip_count'] = df['Position'].apply(lambda x: sum(abs(p) * (25 - i) for i, p in enumerate(x) if p < 0))
```
